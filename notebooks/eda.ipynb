```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/cleaned_data.csv')
```

```python
total_sales=df["Sales"].sum()
total_profit=df["Profit"].sum()
total_orders=df["Order ID"].nunique()
total_customers=df["Customer ID"].nunique()

print("Total Sales:",round(total_sales, 2))
print("Total Profit:",round(total_profit, 2))
print("Total Orders:",total_orders)
print("Total Customers:",total_customers)
```

```python
# Category-wise sales
category_sales=(df.groupby("Category")["Sales"].sum().sort_values(ascending=False))
category_sales
```

```python
# Category-wise profit
category_profit=(df.groupby("Category")["Profit"].sum().sort_values(ascending=False))
category_profit
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=category_sales.index,
    y=category_sales.values,
    palette="Blues_d"
)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=category_profit.index,
    y=category_profit.values,
    palette="Greens_d"
)
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
sales_percent=(df.groupby("Category")["Sales"].sum().sort_values(ascending=False))
sales_percent=(sales_percent / sales_percent.sum()) * 100
sales_percent.round(2)
```

```python
profit_percent=(df.groupby("Category")["Profit"].sum().sort_values(ascending=False))
profit_percent=(profit_percent / profit_percent.sum()) * 100
profit_percent.round(2)
```

Category Analysis

- Technology contributes the highest sales and profit.
- Furniture contributes high sales but comparatively low profit.
- Office Supplies generates lower sales than Furniture but significantly higher profit.
- Technology appears to be the most profitable business category

```python
region_sales_percent=(df.groupby("Region")["Sales"].sum().sort_values(ascending=False))
region_sales_percent=(region_sales_percent / region_sales_percent.sum()) * 100
region_sales_percent.round(2)
```

```python
region_profit_percent=(df.groupby("Region")["Profit"].sum().sort_values(ascending=False))
region_profit_percent=(region_profit_percent / region_profit_percent.sum()) * 100
region_profit_percent.round(2)
```

Region Analysis

- West is the highest-performing region in both sales and profit.
- East is the second strongest region and shows consistent performance.
- Central generates higher sales than South but earns the lowest profit, indicating possible inefficiencies.
- South has the lowest sales but outperforms Central in profit, suggesting better profitability.


```python
monthly_sales = (df.groupby(["Order Month Number", "Order Month"])["Sales"].sum().reset_index().sort_values("Order Month Number"))
monthly_sales
```

```python
monthly_profit=(df.groupby(["Order Month Number", "Order Month"])["Profit"].sum().reset_index().sort_values("Order Month Number"))
monthly_profit
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(
    x=monthly_sales["Order Month"],
    y=monthly_sales["Sales"],
    marker="o"
)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(
    x=monthly_profit["Order Month"],
    y=monthly_profit["Profit"],
    marker="o"
)
plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```

 Monthly Sales & Profit Analysis
- November recorded the highest sales, while February had the lowest sales.
- December generated the highest profit, whereas January recorded the lowest profit.
- Sales and profit generally increase towards the end of the year, indicating seasonal demand.
- The last quarter (September–December) contributes significantly to the company's performance.
- Although November has the highest sales, December achieves the highest profit, suggesting improved profit margins.
