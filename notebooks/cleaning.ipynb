Initial Data Understanding
- Dataset contains 21 rows and 9994 columns
- Number of missing values:0
- Number of duplicate rows:0
- Date columns currently stored as str(15),float(3) and int(3)
- Order Date and Ship Date are currently stored as string datatype which should be converted to date datatype 
- Numeric metrics include Sales, Quantity, Discount and Profit

```python
df_clean=df.copy()
```

```python
df_clean["Order Date"]=pd.to_datetime(df_clean["Order Date"])
df_clean["Ship Date"]=pd.to_datetime(df_clean["Ship Date"])
```

```python
df_clean.info()
```

```python
df_clean["Order Year"] = df_clean["Order Date"].dt.year
df_clean[["Order Date","Order Year"]].head()
```

```python
df_clean["Order Month"] = df_clean["Order Date"].dt.month_name()
df_clean[["Order Date","Order Month"]].head()
```

```python
df_clean["Shipping Days"] = (df_clean["Ship Date"] - df_clean["Order Date"]).dt.days
df_clean[["Order Date","Ship Date","Shipping Days"]].head()
```
